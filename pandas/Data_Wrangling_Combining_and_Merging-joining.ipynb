```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['a', 'b', 'c', 'd', 'e'],'A1': range(5), 'B1':range(5,10)})
df2 = pd.DataFrame({'key': ['a', 'b', 'c'], 'A2': range(3), 'B2':range(3,6)})
```

```python
df1
```

```python
df2
```

## marge()

### how = 'inner'

```python
pd.merge(df1, df2, how='inner', on='key') # only common value
```

### how ='outer'

```python
pd.merge(df1, df2, how='outer', on='key') # common and uncommon
```

### how = 'left'

```python
pd.merge(df1, df2, how='left', on='key') # depend by left table value
```

### how = 'right'

```python
pd.merge(df1, df2, how='right', on='key') # depend by right table value
```

## Merging example with two key (key1, key2) columns -- little complicated!

```python
left = pd.DataFrame({'key1': ['a', 'a', 'b', 'c'],
                     'key2': ['a', 'b', 'a', 'b'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['a', 'b', 'b', 'c'],
                      'key2': ['a', 'b', 'a', 'a'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how='inner', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

## Concatenation

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4,5,6,7]) 
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1,df2],axis=1) # axis = 1/columns
```
